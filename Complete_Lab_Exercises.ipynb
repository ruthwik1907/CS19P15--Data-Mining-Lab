# **Loading a Dataset and Visualising the Data**

```python
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
import pandas as pd
```

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = iris.target_names[iris.target]
```

```python
sns.pairplot(df, hue='species', height=2)
plt.show()
```

# Classification using Classifier **Algorithm**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2)
```

```python
clf = RandomForestClassifier()
clf.fit(X_train, y_train)
```

```python
preds = clf.predict(X_test)
print(f"Accuracy: {accuracy_score(y_test, preds):.2%}")
```

# **Clustering** using Clustering Algorithm

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, n_init='auto')
clusters = kmeans.fit_predict(iris.data)
```

```python
plt.scatter(iris.data[:, 0], iris.data[:, 1], c=clusters, cmap='viridis')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.title('K-Means Clustering')
plt.show()
```

# Approriate Rules based **Appriori Algorithm**

```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
dataset = [['Milk', 'Onion', 'Nutmeg', 'Eggs', 'Yogurt'],
           ['Dill', 'Onion', 'Nutmeg', 'Eggs', 'Yogurt'],
           ['Milk', 'Apple', 'Eggs'],
           ['Milk', 'Unicorn', 'Corn', 'Yogurt'],
           ['Corn', 'Onion', 'Onion', 'Ice cream', 'Eggs']]
```

```python
te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)

# Apply Apriori: Find items with at least 60% support
frequent_itemsets = apriori(df, min_support=0.6, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
print(rules[['antecedents', 'consequents', 'support', 'confidence']])
```

# **Data Wrangling & Manipulation (Pandas)**

```python
import pandas as pd

# 1. Create/Load Data
df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'Alice'],
    'Age': [25, None, 30, 25],
    'Salary': [50000, 60000, 70000, 50000]
})

# 2. Drop duplicates and fill missing values
df = df.drop_duplicates().fillna({'Age': df['Age'].mean()})

# 3. Filtering: Get people older than 26
seniors = df[df['Age'] > 26]

# 4. Aggregation: Average salary by Name
avg_salary = df.groupby('Name')['Salary'].mean()

print("\nGrouped Salary:\n", avg_salary)
```

# **Linear Regression Algorithm**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([3.1, 5.2, 6.9, 9.1, 11.0])
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
plt.scatter(X, y, color='blue') # Actual data
plt.plot(X, model.predict(X), color='red') # Regression line
plt.title(f"Equation: y = {model.coef_[0]:.2f}x + {model.intercept_:.2f}")
plt.show()
```
